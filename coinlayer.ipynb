```python
from yahooquery import Screener

s = Screener()
data = s.get_screeners('all_cryptocurrencies_us', count=250)

# data is in the quotes key
dicts = data['all_cryptocurrencies_us']['quotes']
symbols = [d['symbol'] for d in dicts]
symbols
```

```python
syms = []
for symbol in symbols:
  sym = symbol[:-4]
  syms.append(sym)
  
print(syms)
```

```python
import requests
import pandas as pd

def get_crypto_prices(symbol, start, end):
    api_key = '2740dfa22960bea6261577e6439e3702'
    series = pd.date_range(start, end)
    dates = []
    for i in range(len(series)):
        dates.append(str(series[i]))
    date_series = []
    for i in dates:
        date_series.append(i[:10])
    prices = []
    for date in date_series:
        try:
            api_url = f'http://api.coinlayer.com/{date}&symbols={symbol}?access_key={api_key}'
            raw = requests.get(api_url).json()
            val = []
            val.append(raw['rates'])
            price = val[0][f'{symbol}']
            prices.append(price)
        except:
            prices.append('NaN')
    df = pd.DataFrame(columns = ['date', 'price'])
    df['date'] = series
    df['price'] = prices
    return df


confirmed =[]
for sym in syms:
  try:
    df = get_crypto_prices('BCH', '2012-01-01', '2021-04-27')
  except:
    pass
  try:
    x = df.iloc[-1]['Close']*10 
    y = df.iloc[-31]['Close']
    if x <= y:
      confirmed.append(sym)
      #hist_df.to_csv('/content/drive/MyDrive/crypto/Binance/'+symbol+'.csv')
      print(sym+' is done')
    else:
      print('pass '+ sym)
      pass
  except IndexError:
        pass

```
