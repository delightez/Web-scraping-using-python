```python
from yahooquery import Screener

s = Screener()
data = s.get_screeners('all_cryptocurrencies_us', count=250)

# data is in the quotes key
dicts = data['all_cryptocurrencies_us']['quotes']
symbols = [d['symbol'] for d in dicts]
symbols
```

```python
syms = []
for symbol in symbols:
  sym = symbol[:-4]
  syms.append(sym)
  
print(syms)
```

```python
import requests

url = 'https://web-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'
list_of_symbols= []
for start in range(1, 20000, 5000):

    params = {
        'start': start,
        'limit': 5000,
    }

    r = requests.get(url, params=params)
    data = r.json()
    
    for number, item in enumerate(data['data']):
        list_of_symbols.append(item['symbol'])
print(len(list_of_symbols))
```

```python
from cryptocmd import CmcScraper
confirmed =[]
for sym in list_of_symbols:
  try:
    # initialise scraper without time interval
    scraper = CmcScraper(sym)

    # get raw data as list of list
    headers, data = scraper.get_data()

    # get data in a json format
    xrp_json_data = scraper.get_data("json")

    # export the data as csv file, you can also pass optional `name` parameter
    scraper.export("csv", name="xrp_all_time")

    # Pandas dataFrame for the same data
    df = scraper.get_dataframe()
  except:
    pass
  try:
    x = df.iloc[-1]['Close']*10 
    y = df.iloc[-31]['Close']
    if x <= y:
      confirmed.append(sym)
      df = df.tail(31)
      x = df.iloc[-1]['Close']
      y = df.iloc[-31]['Close']
      
      #hist_df.to_csv('/content/drive/MyDrive/crypto/Binance/'+symbol+'.csv')
      print(sym+' is done')
    else:
      print('pass '+ sym)
      pass
  except IndexError:
        pass
```

```python
from cryptocmd import CmcScraper
symrow =[]
fdrow = []
ldrow = []
fcrow = []
lcrow = []
for sym in list_of_symbols:
  try:
    # initialise scraper without time interval
    scraper = CmcScraper(sym)

    # get raw data as list of list
    headers, data = scraper.get_data()

    # get data in a json format
    xrp_json_data = scraper.get_data("json")

    # export the data as csv file, you can also pass optional `name` parameter
    scraper.export("csv", name="xrp_all_time")

    # Pandas dataFrame for the same data
    df = scraper.get_dataframe()
    firstclose = df.iloc[-1]['Close']
    fcrow.append(firstclose)
    lastclose = df.iloc[-31]['Close']
    lcrow.append(lastclose)
    opendate =  df.iloc[-1]['Date']
    fdrow.append(opendate)
    onemonthdate =  df.iloc[-31]['Date']
    ldrow.append(onemonthdate)
    symrow.append(sym)
    print(sym)



  except:
    pass
 
```
